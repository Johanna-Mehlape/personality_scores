The heat graph above shows us the correlation of our data(relative intensity). Therefore the darker the colour/ more intense the colour means that there is more of negative correlation between the values compared.

i.e we could easily tell that the lowest value here is 0.059 which almost looks as if it is black if not black, therefore it means that between the IPIP_Extraversion and the IPIP_Conscientiousness there is less correlation although it is still positive.
